```python
import pandas as pd
import ast

# Read the CSV files
courses_df = pd.read_csv('sample_courses.csv')
instructors_df = pd.read_csv('sample_instructors.csv')

print("Courses DataFrame:")
print(courses_df.head())
print("\nInstructors DataFrame:")
print(instructors_df.head())

# Create a mapping of course to instructor IDs
course_instructor_mapping = {}

# Process each instructor
for index, row in instructors_df.iterrows():
    instructor_id = row['instructor_id']
    qualified_courses = row['qualified_courses']
    
    # Split the qualified courses by comma
    if pd.notna(qualified_courses):
        courses_list = [course.strip() for course in qualified_courses.split(',')]
        
        # Add instructor ID to each course
        for course in courses_list:
            if course not in course_instructor_mapping:
                course_instructor_mapping[course] = []
            course_instructor_mapping[course].append(instructor_id)

print("\nCourse to Instructor mapping:")
for course, instructors in course_instructor_mapping.items():
    print(f"{course}: {instructors}")
```

```python
# Function to get qualified instructor IDs for a course
def get_qualified_instructors(course_id):
    if course_id in course_instructor_mapping:
        return ', '.join(map(str, course_instructor_mapping[course_id]))
    else:
        return ''

# Update the courses DataFrame with qualified instructor IDs
courses_df['qualified_instructor_ids'] = courses_df['course_id'].apply(get_qualified_instructors)

print("Updated Courses DataFrame:")
print(courses_df.head(20))

# Save the updated DataFrame to a new CSV file
courses_df.to_csv('updated_sample_courses.csv', index=False)
print("\nUpdated courses saved to 'updated_sample_courses.csv'")

# Show some statistics
print(f"\nTotal courses: {len(courses_df)}")
print(f"Courses with qualified instructors: {len(courses_df[courses_df['qualified_instructor_ids'] != ''])}")
print(f"Courses without qualified instructors: {len(courses_df[courses_df['qualified_instructor_ids'] == ''])}")
```

```python
# Show courses that have qualified instructors
print("\nCourses with qualified instructors:")
courses_with_instructors = courses_df[courses_df['qualified_instructor_ids'] != '']
for index, row in courses_with_instructors.iterrows():
    print(f"{row['course_id']}: {row['name']} -> Instructors: {row['qualified_instructor_ids']}")

print("\n" + "="*50)
print("Courses without qualified instructors:")
courses_without_instructors = courses_df[courses_df['qualified_instructor_ids'] == '']
for index, row in courses_without_instructors.iterrows():
    print(f"{row['course_id']}: {row['name']}")

print("\n" + "="*50)
print("Summary of instructor qualifications:")
instructor_course_count = {}
for course, instructors in course_instructor_mapping.items():
    for instructor_id in instructors:
        if instructor_id not in instructor_course_count:
            instructor_course_count[instructor_id] = 0
        instructor_course_count[instructor_id] += 1

print("Instructor ID -> Number of courses they can teach:")
for instructor_id, count in sorted(instructor_course_count.items()):
    instructor_name = instructors_df[instructors_df['instructor_id'] == instructor_id]['instructor_name'].iloc[0]
    print(f"ID {instructor_id} ({instructor_name}): {count} courses")
```
